Add a new column that categorizes employees based on their performance rating: 

```python
value=[]                                             ###creating performace _referance
for i in employee_data['Rating']:
    if i>=4.5 :
        value.append("EXCELLENT")
    elif i<4.5 and i>=4.0:
        value.append("GOOD")
    else :value.append("AVERAGE")
```

```python
employee_data['performance_remarks']=pd.Series(value)              ## adding column
```

Modify the dataset by adding a new column, Tax, which deducts 10% of the 
Salary. 

```python
per=[]
for i in employee_data['Salary'] :
    per.append(i/10)
```

```python
employee_data['TAX']=pd.Series(per)
```

```python
employee_data['GROSS_SALARY']=pd.Series(employee_data['Salary']-employee_data['TAX'])    ##salary after tax deduction
```
