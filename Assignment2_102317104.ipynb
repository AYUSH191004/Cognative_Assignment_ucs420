```python
type(mytuple)
```

Identify the highest score and its index in the tuple.

```python
max_num=max(mytuple)                                       ###max_score in tuple
for i in range(0,len(mytuple)):                                      ##iterating through tuple to get index
    if mytuple[i]==max_num:
        print(f"MAX NUMBER IN TUPLE : {mytuple[i] } at INDEX  {i}")    ##desired output
```

Find the lowest score and count how many times it appears.

```python
min_score=min(mytuple)            ##find min_score in tuple
count=0                                 ## count of min_score in tuple
print(min_score)
for i in range(0,len(mytuple)):
    if mytuple[i]==min_score:
        count+=1

print(f"NUMBER OF TIMES LEAST SCORE APPEARS IN TUPLE : {count}")        ##desired output
```

Reverse the tuple and return it as a list.

```python
l_tuple=[]    ##create a empty list
for i in mytuple:         ##append elements in list sequence wise
    l_tuple.append(i)

l_tuple.reverse()       ##reverse the list 
print(f"REVERSED TUPLE: {l_tuple}")    ##desired output
```

Consider the following two sets, A and B, represenƟng scores of two teams in mulƟple
matches. A = {34, 56, 78, 90} and B = {78, 45, 90, 23}

```python
team_A={34, 56, 78, 90}
team_b={78, 45, 90, 23}
```

Find the unique scores achieved by both teams (union of sets).

```python
unique_score=team_A.union(team_b)   ##union of two sets    
```

```python
unique_score
```

IdenƟfy the scores that are common to both teams (intersecƟon of sets).

```python
common_score=team_A.intersection(team_b)   ##intersection of two teams  
```

```python
common_score
```

Find the scores that are exclusive to each team (symmetric difference).

```python
exclusive_score=team_A.symmetric_difference(team_b)   ##exclusive score of two teams
```

```python
exclusive_score
```

Check if the scores of team A are a subset of team B, and if team B's scores are
a superset of team A.

```python
def subset(s1,s2):   ##function to check  subset
    if s1.issubset(s2):
        return True
    else:
        return False
    
subset(team_A,team_b)   ##desired output
```

```python
def superset(s1,s2):   ##function to check  superset
    if s1.issubset(s2):
        return True
    else:
        return False
    
superset(team_b,team_A)   ##desired output
```
